```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
df.sample(4)
```

```python
df['Survived'].value_counts()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(
    x=df['Survived'].value_counts().index,
    y=df['Survived'].value_counts()
)

plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(data=df, x='Pclass', hue='Survived')

plt.title('Sobrevivientes por clase de pasajero')
plt.xlabel('Clase de pasajero')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sí sobrevivió'])
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(data=df, x='Sex', hue='Survived')

plt.title('Sobrevivientes por género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sí sobrevivió'])
plt.show()
```

```python
plt.figure(figsize=(12, 8))

color = ['#240750', '#344C64']

sns.countplot(data=df, x='Embarked', hue='Survived', palette=color)

plt.title('Supervivencia por puerto de embarque')
plt.xlabel('Puerto de embarque')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sí sobrevivió'])
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.histplot(df, x='Age', kde=True)

plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Cantidad')

plt.axvline(df['Age'].mean(), color='red', label='Media')
plt.axvline(df['Age'].median(), color='green', label='Mediana')
plt.axvline(df['Age'].mode()[0], color='black', label='Moda')

plt.legend()
plt.show()

print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```

```python
plt.figure(figsize=(12, 8))

sns.heatmap(
    df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(),
    annot=True,
    cbar=True,
    cmap='coolwarm',
    cbar_kws={'label': 'Nivel de correlación'}
)

plt.title('Correlación de variables numéricas')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(data=df, x='Sex', hue='Survived')

plt.title('Cantidad de sobrevivientes por género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sí sobrevivió'])

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot(data=df, x='Embarked', hue='Survived')

plt.title('Cantidad de sobrevivientes por puerto de embarque')
plt.xlabel('Puerto de embarque')
plt.ylabel('Cantidad')
plt.legend(title='Supervivencia', labels=['No sobrevivió', 'Sí sobrevivió'])

plt.show()
```

```python
bins = [0, 5, 12, 18, 30, 60, 100]
labels = ['Infante', 'Niño', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']

df['Grupo_Edad'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

tasa_supervivencia_edad = df.groupby('Grupo_Edad', observed=True)['Survived'].mean()

plt.figure(figsize=(12, 8))

sns.barplot(
    x=tasa_supervivencia_edad.index,
    y=tasa_supervivencia_edad.values
)

plt.title('Tasa de supervivencia por grupo de edad')
plt.xlabel('Grupo de edad')
plt.ylabel('Tasa de supervivencia')

plt.show()
```

En la gráfica se observa que el grupo de personas mayores tuvo una menor tasa de supervivencia en comparación con otros grupos de edad. Esto sugiere que la edad pudo influir en la probabilidad de sobrevivir, aunque también podrían intervenir otros factores como el género o la clase del pasajero.
